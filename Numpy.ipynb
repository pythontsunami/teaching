# Numpy



The fundamental package for scientific computing with python. It provides a high performance multidimensional array object and tools to work with these arrays.

```python
import numpy as np
```

## Arrays

A numpy array is a grid of values, all of the same type, and is indexed by a tuple of nonnegative integers.

```python
from IPython.display import Image
Image("images/numpy_arrays.png")
```

### Creating arrays

```python
x = np.array([3,1,-2])
x
```

Create array of zeros

```python
x = np.zeros(10)
x
```

```python
np.zeros((2,3))
```

Create array of ones

```python
x = np.ones(4)
x
```

```python
np.ones((2,4))
```

Create a constant array (i.e. all entries are the same)

```python
np.full(10, 2)
```

```python
np.full((2,5),2)
```

Create array with random values (between 0 and 1)

```python
np.random.random(10)
```

Create array with all numbers from -5 to 5

```python
np.arange(-5,5)
```

```python
np.arange(-5,5)
```

Create an array with evenly spaced numbers

```python
np.linspace(1, 10, num=10)
```

Create an array with numbers spaced evenly on a log scale.

```python
np.logspace(1, 10, num=10, base=10)
```

#### 2. Calculate the standard deviation  without using np.std() and np.mean()

$std=\sqrt{\frac{\sum_{i=1}^{N}\left(x_{i}-\bar{x}\right)}{N-1}}$

```python
mean = np.sum(a)/len(a)
mean
```

```python
np.mean(a)
```

```python
np.sqrt(np.sum((a-mean)**2)/len(a))
```
