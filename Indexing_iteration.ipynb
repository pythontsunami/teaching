# Indexing and iteration

```python
import numpy as np
```

## Indexing

```python
array_a = np.array([3,4,5,6,7,202,11,240,1])
array_a
```

Get third element in the array

```python
array_a[2]
```

Get first element in the array

```python
array_a[0]
```

Get last element in the array 

```python
array_a[-1]
```

Get next to last element in the array

```python
array_a[-2]
```

Get all entries except for the first two

```python
array_a[2:]
```

Get all entries except for the last three

```python
array_a[:-3]
```

Get every other entry

```python
array_a[::2]
```

```python
array_a[0:len(array_a):2]
```

Get every other entry starting in the second one

```python
array_a[1:len(array_a):2]
```

Changing array entries

```python
array_a[2] = 10001
array_a
```

In a 2D array (matrix)

```python
array_b = np.array([[2,3,5], [35,1,5]])
array_b
```

```python
array_b[1,2]
```

Same thing can be done with python lists

```python
list_a = list(array_a)
list_a
```

```python
list_a[2]
```

```python
list_a[2] = 2000
list_a
```

```python
list_b = [[2,3,5], [35,1,5]]
list_b
```

```python
list_b[0][1]
```

# Iteration

```python
n = len(list_a)
n
```

### While loop

```python
i = 0
while i < n:
    print(i, array_a[i], list_a[i])
    i = i + 1
```

```python
i = 3
while i < n:
    print(i, array_a[i], list_a[i])
    i = i + 1
```

### For loop

```python
for i in range(n):
    print(i, array_a[i], list_a[i])
```

```python
list(range(n))
```

```python
for i in range(3, n):
    print(i, array_a[i], list_a[i])
```

```python
list(range(3, n))
```

```python
new_list = []
for i in range(n):
    new_list.append(list_a[i] ** 2)
```

```python
new_list
```
